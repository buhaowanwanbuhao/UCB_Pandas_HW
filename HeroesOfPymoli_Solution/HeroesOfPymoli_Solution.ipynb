```python
import pandas as pd

# Import the dataset
purchase_dataframe=pd.read_csv("purchase_data.csv")
purchase_dataframe
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_dataframe["SN"].value_counts())

# total_players value into dataframe and save in Total_Players
Total_Players=pd.DataFrame({"Total Players":[total_players]})
Total_Players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique Items
number_of_unique_items = purchase_dataframe["Item ID"].nunique()
#Avg Purchase Price
average_purchase_price = purchase_dataframe["Price"].mean()
#Total Number of Purchases
total_number_of_purchase = len(purchase_dataframe.index)
#Total Revenue
Total_Revenue = purchase_dataframe["Price"].sum()

purchase_total_dataframe = pd.DataFrame.from_dict({"Number of unique items":[number_of_unique_items],"Average Price":[average_purchase_price],"Number of Purchases":[total_number_of_purchase],"Total Revenue":[Total_Revenue]}, orient='columns')
purchase_total_dataframe["Total Revenue"] = purchase_total_dataframe["Total Revenue"].map("$ {:,.2f}".format)
purchase_total_dataframe["Average Price"] = purchase_total_dataframe["Average Price"].map("$ {:,.2f}".format)
purchase_total_dataframe
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics
#Percentage and Count of Male Players
gender_dataframe_heroes = pd.DataFrame(purchase_dataframe.drop_duplicates(["SN"]))
gender_demographics_group = gender_dataframe_heroes.groupby(['Gender'])
gender_dataframe = pd.DataFrame({"Total Count":gender_demographics_group["Gender"].count(),"Percentage of players":gender_demographics_group["Gender"].count()/total_players*100})
gender_dataframe["Percentage of players"] = gender_dataframe["Percentage of players"].map("{:.2f}%".format)
gender_dataframe.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_gender = purchase_dataframe.groupby(['Gender'])
gender_purchase_dataframe = pd.DataFrame({"Purchase Count":purchase_gender["Price"].count(), "Average Purchase Price":purchase_gender["Price"].mean(),"Total Purchase Value":purchase_gender["Price"].sum(), "Normalized Totals":purchase_gender["Price"].sum()/gender_dataframe["Total Count"]})
gender_purchase_dataframe["Average Purchase Price"] = gender_purchase_dataframe["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_dataframe["Total Purchase Value"] = gender_purchase_dataframe["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_dataframe["Normalized Totals"] = gender_purchase_dataframe["Normalized Totals"].map("${:,.2f}".format)
gender_purchase_dataframe.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Segment and sort age values into bins established above
purchase_dataframe["Age Group"] = pd.cut(purchase_dataframe["Age"],age_bins, labels=group_names)

# Create new data frame with the added "Age Group" and group it
age_grouped = purchase_dataframe.groupby("Age Group")

# Count total players by age category
total_count_age = age_grouped["SN"].nunique()

# Calculate percentages by age category 
percentage_by_age = (total_count_age/total_players) * 100

# Create data frame with obtained values
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": total_count_age})

# Format the data frame with no index name in the corner
age_demographics.index.name = None

# Format percentage with two decimal places 
age_demographics.style.format({"Percentage of Players":"{:,.2f}"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Dataframe_Age = pd.DataFrame(purchase_dataframe)
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]

Dataframe_Age_demographic_group = Dataframe_Age.groupby(pd.cut(Dataframe_Age["Age"], bins, labels=group_names))

# Create the names for the four bins
demographic_age_dataframe = pd.DataFrame({"Purchase Count":Dataframe_Age_demographic_group["Price"].count(), "Average Purchase Price":Dataframe_Age_demographic_group["Price"].mean(),"Total Purchase Value":Dataframe_Age_demographic_group["Price"].sum(),"Normalized Totals":Dataframe_Age_demographic_group["Price"].sum()/Dataframe_Age_demographic_group["SN"].nunique()})
demographic_age_dataframe["Average Purchase Price"] = demographic_age_dataframe["Average Purchase Price"].map("${:.2f}".format)
demographic_age_dataframe["Total Purchase Value"] = demographic_age_dataframe["Total Purchase Value"].map("${:,.2f}".format)
demographic_age_dataframe["Normalized Totals"] = demographic_age_dataframe["Normalized Totals"].map("${:,.2f}".format)
demographic_age_dataframe
```

## Most Popular Items

```python
top_spenders_group = purchase_dataframe.groupby(['SN'])
top_spenders_group_dataframe = pd.DataFrame({"Purchase Count":top_spenders_group["Price"].count(), "Average Purchase Price":top_spenders_group["Price"].mean(),"Total Purchase Value":top_spenders_group["Price"].sum()})
top_spenders_group_dataframe= top_spenders_group_dataframe.sort_values("Total Purchase Value", ascending=False)
top_spenders_group_dataframe["Average Purchase Price"] = top_spenders_group_dataframe["Average Purchase Price"].map("${:.2f}".format)
top_spenders_group_dataframe["Total Purchase Value"] = top_spenders_group_dataframe["Total Purchase Value"].map("${:,.2f}".format)
top_spenders_group_dataframe.head()
```

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popitems_group = purchase_dataframe.groupby(['Item ID','Item Name'])
popitems_group_dataframe = pd.DataFrame({"Purchase Count":popitems_group["Price"].count(), "Item Price":popitems_group["Price"].mean(),"Total Purchase Value":popitems_group["Price"].sum()})
popitems_group_dataframe = popitems_group_dataframe.sort_values("Purchase Count", ascending=False)
popitems_group_dataframe["Item Price"] = popitems_group_dataframe["Item Price"].map("${:.2f}".format)
popitems_group_dataframe["Total Purchase Value"] = popitems_group_dataframe["Total Purchase Value"].map("${:,.2f}".format)
popitems_group_dataframe.head()
```
